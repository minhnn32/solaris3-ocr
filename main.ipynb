```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Pre-processing

```python
myImage = cv2.imread('images/test_screenshot.png', cv2.IMREAD_UNCHANGED)
grayImg = cv2.cvtColor(myImage, cv2.COLOR_RGB2GRAY)
plt.imshow(grayImg, cmap='gray')
```

```python
ret, thresh = cv2.threshold(grayImg, 0, 255, cv2.THRESH_OTSU+cv2.THRESH_BINARY, grayImg)
plt.imshow(thresh, cmap='gray')
```

```python
horizontal_hist = np.sum(grayImg, axis=1, keepdims=True)/255

# draw the horizontal histogram along with the original image on the same plot

fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].imshow(grayImg, cmap='gray', aspect='auto')
ax[0].set_title('Original Image')
# plot using same height as the original image
ax[1].plot(horizontal_hist, range(grayImg.shape[0]), color='red')
ax[1].set_xlim([0, grayImg.shape[1]])
ax[1].set_ylim([0, grayImg.shape[0]])
ax[1].invert_yaxis()
ax[1].set_title('Horizontal Histogram')
plt.show()
```

```python
# using histograms to segment the image

# take derivative of the histogram
horizontal_hist_diff = np.diff(horizontal_hist, axis=0, prepend=0)

# draw the horizontal histogram along with the original image on the same plot
fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].imshow(grayImg, cmap='gray', aspect='auto')
ax[0].set_title('Original Image')
# plot using same height as the original image
ax[1].plot(horizontal_hist_diff, range(grayImg.shape[0]), color='red')
ax[1].set_xlim([0, grayImg.shape[1]])
ax[1].set_ylim([0, grayImg.shape[0]])
ax[1].invert_yaxis()
ax[1].set_title('Horizontal Histogram')
# plot the peaks
plt.show()
```
